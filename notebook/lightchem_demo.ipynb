```python
## lightchem demo
# Author: Haozhen Wu
# Date: 12/5/2016

import sys
base_dir = '/Users/alan/Desktop/study/SmallMolecular/github/'
sys.path.append(base_dir + "lightchem/lightchem/load/")
sys.path.append(base_dir + "lightchem/lightchem/envir/")
sys.path.append(base_dir + "lightchem/lightchem/fold/")
sys.path.append(base_dir + "lightchem/lightchem/data/")
sys.path.append(base_dir + "lightchem/lightchem/model/")
sys.path.append(base_dir + "lightchem/lightchem/eval/")
from sklearn import metrics
import create_dir
import load
import fold
import xgb_data
import xgb_eval
import eval_testset
import first_layer_model
import second_layer_model
```

```python
base_data_dir = '/Users/alan/Desktop/study/SmallMolecular/pcba128_multitask_data/'
binaryLabel_dir = 'pcba128_mmtn_canon_ecfp1024.csv'
label_colname = 'pcba-aid411'
binaryLabel = load.readData(base_data_dir + binaryLabel_dir,label_colname)
binaryLabel.read()
```

```python
# Get features and label 
X_data = binaryLabel.features()
y_data = binaryLabel.label()
```

```python
X_data
```

```python
y_data
```

```python
# Generate stratified k-folds
number_fold = 4
myfold = fold.fold(X_data,y_data,number_fold)
myfold = myfold.generate_skfolds()
```

```python
# cross validation fold index
myfold.head()
```

```python
# xgbData is lightchem's default data object. 
# First dataset to use
binaryLabel_data = xgb_data.xgbData(myfold,X_data,y_data)
binaryLabel_data.build()
```

```python
continusLabel_dir = 'pcba128_canon_ecfp1024_logac50.csv'
label_colname = 'aid411_logAC50'
continusLabel = load.readData(base_data_dir + continusLabel_dir,label_colname)
continusLabel.read()
```

```python
X_data2 = continusLabel.features()
y_data2 = continusLabel.label()
```

```python
X_data2
```

```python
y_data2
```

```python
# In virtual screening, we often have many different kinds of input features and label.
# For example, we can use raw score(score that convert to 1/0 label) to train a regression and later ensemble together.
# Or we can have different featurization method for molecules. Using all of them in first layer models and ensemble
# them in second layer models.

# Continus data use same fold generated before.
continusLabel_data = xgb_data.xgbData(myfold,X_data2,y_data2)
continusLabel_data.build()
```

```python
# model based on binary data
evaluation_metric_name = 'ROCAUC'
model_type = 'GbtreeLogistic' # Gradient boosting tree for logistic regression
unique_name = 'layer1_GbtreeLogistic'
model1 = first_layer_model.firstLayerModel(binaryLabel_data,
        evaluation_metric_name,model_type,unique_name)
```

```python
# run cross validation. lightchem use cross validation models to make prediction
model1.xgb_cv()
```

```python
# Cross validation scores
model1.cv_score()
```

```python
# Generate holdout(out of fold) prediction. Will be used for second layer models
model1.generate_holdout_pred()
```

```python
# Holdout prediction
model1.get_holdout()
```

```python
# model based on Continuous data
evaluation_metric_name = 'ROCAUC'
model_type = 'GbtreeRegression'
unique_name = 'layer1_GbtreeRegression'
model2 = first_layer_model.firstLayerModel(continusLabel_data,
        evaluation_metric_name,model_type,unique_name)
```

```python
model2.xgb_cv()
```

```python
# Cross validation scores
model2.cv_score()
```

```python
# Generate holdout(out of fold) prediction. Will be used for second layer models
model2.generate_holdout_pred()
```

```python
# Holdout prediction
model2.get_holdout()
```

```python
# use label from binary data to train layer2 models
list_l1model = [model1,model2]
evaluation_metric_name = 'ROCAUC'
model_type = 'GbtreeLogistic'
unique_name = 'layer2_GbtreeLogistic'
l2model = second_layer_model.secondLayerModel(binaryLabel_data,list_l1model,
            evaluation_metric_name,model_type,unique_name)
```

```python
# Create second layer data
l2model.second_layer_data()
```

```python
# Cross validation. Use same fold as first layer model
l2model.xgb_cv()
```

```python
# Cross validation scores
l2model.cv_score()
```

```python
list_data = [binaryLabel_data.get_dtest()]
test_label = binaryLabel_data.get_testLabel()
evaluation_metric_name = 'ROCAUC'
# Test score
eval_testset.eval_testset(model1,list_data,test_label,evaluation_metric_name)
```

```python
list_data = [continusLabel_data.get_dtest()]
test_label = binaryLabel_data.get_testLabel()
evaluation_metric_name = 'ROCAUC'
# Test score
eval_testset.eval_testset(model2,list_data,test_label,evaluation_metric_name)
```

```python
list_data = [binaryLabel_data.get_dtest(),continusLabel_data.get_dtest()]
test_label = binaryLabel_data.get_testLabel()
evaluation_metric_name = 'ROCAUC'
# Test score
eval_testset.eval_testset(l2model,list_data,test_label,evaluation_metric_name)
```
